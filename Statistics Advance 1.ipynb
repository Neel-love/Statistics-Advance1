```python
#Qus 3. Write a python function to calculate the probabilty distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x: outcome (0 or 1)
    - p: probability of success (0 <= p <= 1)

    Returns:
    - Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    
    return p if x == 1 else (1 - p)
```

```python
#Qus 6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters of the discrete uniform distribution
a = 1 
b = 6

# Create the distribution using scipy.stats.randint
dist = randint(a, b + 1)

# Generate x values (a bit wider range for better CDF visualization)
x = np.arange(a - 2, b + 3)

# Calculate CDF values
cdf_values = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf_values, where='post', color='blue', label='CDF')
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform [{a}, {b}]')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(x)
plt.yticks(np.linspace(0, 1, b - a + 2))
plt.legend()
plt.show()
```

```python
#Qus 9. Write a python function to calculate Z-xcores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Generate a sample dataset
num_samples = 1000
data = np.random.normal(loc=50, scale=10, size=num_samples)  # Normal distribution with mean=50, std=10

# Compute Z-scores
z_scores = calculate_z_scores(data)

# Plot the histogram of Z-scores
plt.hist(z_scores, bins=50, density=True, edgecolor='black', alpha=0.7)
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.title('Histogram of Z-scores')
plt.show()
```

```python
#Qus 15. Wite a python program to calulate the Z-score for a given data point and compare it to a stanard normal distribution .
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score

# Generate a sample dataset
num_samples = 1000
data = np.random.normal(loc=50, scale=10, size=num_samples)  # Normal distribution with mean=50, std=10

# Define a data point for comparison
data_point = 60
z_score = calculate_z_score(data_point, data)

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Mark the Z-score on the plot
plt.axvline(z_score, color='red', linestyle='dashed', label=f'Z-score = {z_score:.2f}')

plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Comparison of Z-score to Standard Normal Distribution')
plt.legend()
plt.grid()
plt.show()
```

```python
#Qus 16. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

# Sample dataset (random values)
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=40)  # Mean=50, Std=10, Sample size=40

# Given population parameters
population_mean = 50
population_std = 10  # Known population standard deviation
sample_size = len(sample_data)

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Set significance level (alpha)
alpha = 0.05

# Decision rule
if p_value < alpha:
    print(f"Reject the null hypothesis (Z = {z_score:.2f}, p-value = {p_value:.5f})")
else:
    print(f"Fail to reject the null hypothesis (Z = {z_score:.2f}, p-value = {p_value:.5f})")
```

```python
#Qus 20. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

# Generate a sample dataset
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=40)  # Mean=50, Std=10, Sample size=40

# Compute sample mean and standard error
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample standard deviation
sample_size = len(sample_data)
standard_error = sample_std / np.sqrt(sample_size)

# Set confidence level
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for 95% confidence

# Compute margin of error
margin_of_error = z_critical * standard_error

# Compute confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level*100}%): {confidence_interval}")
```

```python
#Qus 25.  Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(42)
dataset1 = np.random.normal(loc=50, scale=10, size=500)  # Mean=50, Std=10
dataset2 = np.random.normal(loc=50, scale=20, size=500)  # Mean=50, Std=20 (higher spread)

# Compute standard deviations
std1 = np.std(dataset1, ddof=1)  # Sample standard deviation
std2 = np.std(dataset2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Visualization: Histograms
plt.figure(figsize=(10, 5))
plt.hist(dataset1, bins=30, alpha=0.6, color='blue', label='Dataset 1 (Std=10)')
plt.hist(dataset2, bins=30, alpha=0.6, color='red', label='Dataset 2 (Std=20)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Standard Deviations')
plt.legend()
plt.grid()
plt.show()

# Visualization: Box Plot
plt.figure(figsize=(7, 5))
plt.boxplot([dataset1, dataset2], labels=['Dataset 1 (Std=10)', 'Dataset 2 (Std=20)'])
plt.title('Box Plot Comparison of Standard Deviations')
plt.ylabel('Value')
plt.grid()
plt.show()
```

```python
#Qus 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Sample size=1000

# Calculate range (max - min)
data_range = np.max(data) - np.min(data)

# Calculate interquartile range (IQR)
q1 = np.percentile(data, 25)  # First quartile (Q1)
q3 = np.percentile(data, 75)  # Third quartile (Q3)
iqr = q3 - q1

# Print results
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#Qus 27. Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample dataset (Normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15, Sample size=1000

# Apply Z-score normalization
mean = np.mean(data)
std_dev = np.std(data)
normalized_data = (data - mean) / std_dev  # Z-score normalization

# Visualization: Before and After
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Original Values')
plt.ylabel('Frequency')
plt.title('Original Dataset')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='red', edgecolor='black')
plt.xlabel('Z-score Normalized Values')
plt.ylabel('Frequency')
plt.title('Normalized Dataset')

plt.tight_layout()
plt.show()
```

```python
#Qus 28. Write a python function to calculate the skewness and kurtosis of  a dataset generated from a normal distribution. 
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    skewness = stats.skew(data)  # Measure asymmetry
    kurtosis = stats.kurtosis(data, fisher=True)  # Measure tail heaviness (Fisher=True means excess kurtosis)
    return skewness, kurtosis

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10, Sample size=1000

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print results
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis (excess): {kurtosis:.2f}")
```
